### 기본 함수

- count(): 텍스트에서 어떤 요소의 개수를 세는 함수
- replace(A, B): A를 B로 바꾸는 함수

- index(): 원하는 문자열의 인덱스를 반환하는 함수

#### count 함수
- 텍스트에서 어떤 요소의 개수를 세는 함수

```python
text2.count('the')
```

#### replace 함수
- 요소를 교체하는 함수
- replace(A, B)
    - 모든 A를 B로 교체

```python
word = 'the'
word = word.replace('t', 'T') # 변수를 다시 담아야 할당이 됨
print(word)
```

#### strip 함수
- 원하는 요소를 제거하는 함수 (기본은 공백)
- strip()
    - 양쪽 제거
- rstrip()
    - 오른쪽 제거
- lstrip()
    - 왼쪽 제거

```python
word = '        \t word'

print(word.lstrip())
```

#### find 함수
- 원하는 문자열을 찾는 함수
- 원하는 문자열이 없는 경우, `-1` 반환
- 제일 먼저 나오는 문자열만 그 대상이 됨

```python
word = 'I have a pen.'

print(word.find('a'))
print(word.find('r'))
```

#### index 함수
- 원하는 문자열의 인덱스 반환
- 원하는 문자열이 없는 경우, `Error` 반환

```python
word = 'I have a pen.'

print(word.index('a'))
print(word.index('r'))
```

#### Frequency Distribution 함수
- 단어가 몇 번 출현했는지(빈도수)를 알려주는 함수
- 텍스트의 성질 알기 위해
    - 단어 개수
    - 단어 빈도수
- `FreqDist(<텍스트>)`
    - 텍스트 내 단어들의 개수를 정리함
- `.most_common(<숫자>)`
    - 가장 빈번한 단어를 숫자의 개수만큼 나타냄
- `.freq(<단어>)`
    - 단어가 텍스트에서 차지하는 비율을 나타냄
- `.plot(<반환할 상위 항목 개수>)`
    - 누적 그래프를 원한다면, `cumulative=True` 추가
- `.hapaxes()`
    - 가장 빈번하지 않은 단어(1번만 나타나는 단어)를 나타냄
- fdist = FreqDist(text1)
    - 대소문자, 첫글자 대문자 등등이 각각 세어짐

```python
fdist = FreqDist(text1)
fdist
```

```python
fdist.most_common(10)
```

```python
fdist['that']
```

```python
fdist.freq('that')
```

```python
fdist.plot(50, cumulative=True)
```

```python
fdist.hapaxes()
```

### 리스트(list)
- 파이썬의 `리스트`는 프로그래밍의 `배열`에 해당
- 하나의 변수에 여러 값을 할당하는 자료형
    - 'example = [1, 2, 3, 4]'
- 대괄호를 사용함
- 내부 요소들의 자료형은 같을 필요 없음
    - 'example = [1, 'yes', True]'
- 순서가 있음
    - 슬라이싱, 인덱싱 가능
- 빈 리스트 선언: set(), []

### 리스트 컴프리핸션(list comprehension)
- 포함형/지능형/축약형 리스트
- 하나의 리스트에서 그걸 포함하여 다른 리스트를 만든다.
- list + for loop
- [A for A' in B]
    - B에 있는 A'에서 A를 가져와라 => 리스트에 담아서 반환
    - A: 내가 원하는 출력 형태
    - A': B의 형태
    - B: 반복 가능한 자료형 (list)
    - 현재 단계에서는 A'를 변형하지 않아도 됨
- 조건식 추가
    - [A for A' in B if <조건식>]

```python
# 기본 형태
colors = ['red', 'blue', 'purple', 'black']

[word for word in colors]
```

```python
print(text1.tokens[:10])

# text1의 토큰 중 처음 10개 단어에서 마지막 알파벳 반환

print([word[-1] for word in text1.tokens[:10]])
```

```python
# text5에서 길이가 15 이상인 단어 추출하기
print([word for word in text5 if  len(word) >= 15])
```

```python
# text3에서 대문자를 소문자로 반환
[word.lower() for word in text3 if word.isupper()]
```

```python
# text1에서 'ableness'로 끝나는 단어를 중복없이, 정렬해서
sorted(set([word for word in text1 if word.endswith('ableness')]))
```

```python
# text4에서 'gnt'가 포함된 단어들을 소문자로 리스트 반환
[word.lower() for word in set(text4) if 'gnt' in word]
set([word.lower() for word in text4 if 'gnt' in word])
```

```python
# 대소문자 구분 없이 freq dist 만들기
# ex) the 15,000 The 600 => the 21,000

fdist2 = FreqDist([word.lower() for word in text1])
print(fdist2['the'])

print(fdist2['The'])
```

text2에서 길이가 3자 이하, 빈도수 5 이상

```python
#1
fdist3 = FreqDist(text2)
print([word for word in text2 if len(word) <= 3 and fdist3[word] >= 5])
```

```python
#2: Dictionary 형태인 `FreqDist(text2)`를 사용하는 방법
print([word for (word, count) in FreqDist(text2).items()\
     if (len(word) <= 3) and (count >= 5)])
```

### 리스트 <-> 스트링
- 스트링 -> 리스트
    - split()
    - `str`.split(`separator`)
- 리스트 -> 스트링
    - join()
    - `separator`.join(`str`)

```python
sent = 'I, have, a, car'
sent.split(',')
sent_tokens = [word.strip() for word in sent.split(',')]
```

```python
'-'.join(sent_tokens)
```
