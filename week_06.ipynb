### 2. 여러 변수 할당하기
- packing: 한 변수에 여러 개의 데이터를 할당하는 것
    - tuple, list 자체가 packing
- unpacking: 여러 변수에 하나의 데이터를 할당하고 그것을 각각의 변수로 변환하는 것
- a와 b의 값 교환
    - 일반: `temp1 = a \ temp2 = b \ a = temp2 \ b = temp1`
    - Tuple: `a, b = b, a`

```python
# packing
a = [1, 2, 3, 4]
print(a)

# unpacking
a, b, c, d = 1, 2, 3, 4
print(a, d)
```

```python
name = 'Hwang'

a, b, c, d, e = name
print(a, b, c, d, e)
```

### List 관련 함수

#### 항목을 추가하는 함수
- `<list>.append(<item>)`
    - list의 마지막에 요소를 추가
    - 할당 없이도 할당됨
    - 반환은 따로 해주어야 함
    - list를 추가하면 list 그대로 들어감
- `<list>.insert(<offset>, <item>)`
    - 지정한 오프셋에 요소 추가
- `<list>.extend(<item>)`
    - list에 item을 병합

```python
days = ['월', '화', '수', '목']
days.append('금')

print(days)
```

```python
days = ['월', '화', '목']
days.insert(2, '수')

print(days)
```

```python
# extend
listA = [1, 2, 3]
listB = ['토', '일']

listA.extend(listB)
print('extend:', listA)

# append
listA = [1, 2, 3]
listB = ['토', '일']

listA.append(listB)
print('append:', listA)
```

#### 항목을 삭제하는 함수
- `del <list>[<offset>]`
    - offset에 해당하는 요소 삭제
    - 자리가 보존되지 않고 아이템이 하나씩 앞당겨짐
-  `<list>.remove(<item>)`
    - 특정한 값을 삭제
    - 리스트 중 가장 먼저 나온 하나만 삭제
- `<list>.pop(<index>)`
    - 지정한 인덱스의 요소를 삭제하고 삭제되는 요소를 반환하기
    - default = -1
- `<list>.clear()`
    - 리스트에 있는 모든 요소 삭제

```python
days = ['월', '화', '수']
del days[0]
print(days)
```

```python
days = ['월', '화', '수', '목', '금', '토', '일', '월']
days.remove('월')
print(days)
```

```python
days = ['월', '화', '수', '목', '금', '토', '일']
days.pop()
```

```python
days = ['월', '화', '수', '목', '금', '토', '일']
days.clear()
print(days)
```

#### 항목을 정렬하는 함수
- `<list>.sort()`
    - 리스트를 정렬함
    - 기존 값 자체를 바꿈
    - 반환을 따로 해주어야 함
    - `<list>.sort(reverse=True)`
- `sorted(<item>)`
    - 정렬된 복사본을 반환, 기존 값 유지
    - `sorted(<item>, reverse=True)`
- `<list>.reverse()`
    - 리스트를 뒤집음
    - 기존 값 자체를 바꿈
    - 반환을 따로 해주어야 함
- `list(reversed(<list>))`

```python
woordenlijst = ['greden', 'toegangsprijs', 'volwassenen', 'bewolkt']
```

```python
print(sorted(woordenlijst))
print(woordenlijst)
```

```python
woordenlijst.sort()
print(woordenlijst)
```

```python
b = [1, 2, 6, 43, 4]
print(list(reversed(b)))
print(b)
b.reverse()
print(b)
```

### 할당하기
- 리스트 `a`를 `b = a`로 `b`에 할당하면 b가 바뀌었을 때 `a도 바뀜`
    - 값에 a와 b가 메모리 주소를 참조하기 때문
- shallow copy: a가 변해도 b가 변하지 않게 하는 복사 (1차원)
    - 메모리 주소가 다 다름
    - `a.copy()`
    - `list(a)`
    - `a[:]`
- deep copy: shallow copy보다 더 고정된 복사본을 만드는 복사 (1차원 이상)
    - 하위에 중첩된 리스트 요소가 바뀌는 것을 방지
    - `import copy` \ `copy.deepcopy(<list>)`

```python
a = [0, 1, 2, 3, 4]
```

```python
b = a
b[0] = 7
b
```

```python
a
```

shallow copy

```python
a = [1, 234, 5, 4, 4, 51, 14]
b = a.copy()
c = list(a)
d = a[:]

id(a), id(b), id(c), id(d)
```

deep copy

```python
a = [1, 2, 3, [4, 5, 6]]
b = a.copy()

a[-1][0] = 'x'
print(a)
print(id(a))

print(b)
print(id(b))
```

```python
a = [1, 2, 3, [4, 5, 6]]

import copy
b = copy.deepcopy(a)
print('deep copy b:', b, '\n')

a[-1][-1] = 'x'
print('deep copy a:', a)
print('deep copy b:', b)
print('id a:', id(a))
print('id b:', id(b))
```

### 순회하기
- zip
    - `for <list1>, <list2> in zip(<list1>, <list2>)`

    - 동시적인 개념이기 때문에, `for i in <list1>: for j in <list2>`와는 다름.
- enumerate
    - `for idx, item in enumerate(<list>)`
    - 인덱스도 함께 돌고자 할 때 (list, range)
- range()

zip

```python
a = [1, 2, 3]
b = ['Mon', 'Tue', 'Wed', 'Thu']

for itemA, itemB in zip(a, b):
    print((itemA), (itemB))
```

```python
print(list(zip(a, b)))
print(dict(zip(a, b)))
```

```python
a = ['Mon', 'Tue', 'Wed']

for idx, itemA in enumerate(a):
    print((idx, itemA))
```

## NLTK Corpus

```python
import nltk
nltk.download(['inaugural', 'brown'])
```

```python
brown = nltk.corpus.brown
```

```python
cfd1 = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)

genres = brown.categories()[:5]
modals = ['can', 'could', 'will', 'would', 'may', 'might']
cfd1.tabulate(conditions=genres, samples=modals)
```

genre 별로 modal의 빈도수를 가진 cfd 만들기

```python
cfd2 = nltk.ConditionalFreqDist(
    (genre, modals) # x축과 y축을 바꾸면 그래프도 바뀜
    for genre in brown.categories()
    for modals in ['can', 'could', 'will', 'would', 'may', 'might']
    
    for word in brown.words(categories=genre) # 윗 두 줄을 연결시켜주는 역할
        if word.lower() == modals
)

cfd2.tabulate()
```

```python
cfd2.plot()
```

```python
# 사이즈 키우기

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
cfd2.plot()
```

#### inaugural에서 america, citizen의 빈도수, 연도별로

```python
# 표로

cfd3 = nltk.ConditionalFreqDist(
    (fileid[:-4], target_word)
    for fileid in inaugural.fileids()
    for target_word in ['america', 'citizen']
    
    for word in inaugural.words(fileids=fileid)
        if word.lower().startswith(target_word)
)

cfd3.tabulate()
```

```python
# 그래프로

cfd3 = nltk.ConditionalFreqDist(
    (target_word, fileid[:-4])
    for fileid in inaugural.fileids()
    for target_word in ['america', 'citizen']
    
    for word in inaugural.words(fileids=fileid)
        if word.lower().startswith(target_word)
)

cfd3.plot()
```
